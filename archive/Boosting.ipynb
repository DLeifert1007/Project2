```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import balanced_accuracy_score
```

```python
# import data
df = pd.read_csv('sports_management_encoded.csv')
df.head(10)
```

```python


# Split the features and target
X = df.drop('Sustainability Score', axis=1)
y = df['Sustainability Score']

# Prepare the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Train the Gradient Boosting classifier
# save predictions to a variable
clf1 = GradientBoostingClassifier(random_state=1).fit(X_train, y_train)
prediction_clf1 = clf1.predict(X_test)

# Evaluate the model
print(f'Training Score: {clf1.score(X_train, y_train)}')
print(f'Testing Score: {clf1.score(X_test, y_test)}')

# Train the AdaBoostClassifier
# save predictions to a variable
clf2 = AdaBoostClassifier(random_state=1).fit(X_train, y_train)
prediction_clf2 = clf2.predict(X_test)


# Evaluate the model
print(f'Training Score: {clf2.score(X_train, y_train)}')
print(f'Testing Score: {clf2.score(X_test, y_test)}')
```

```python
# that 
from sklearn.metrics import confusion_matrix, classification_report, balanced_accuracy_score, roc_auc_score
```

```python
# Import SMOTE from imblearn
from imblearn.over_sampling import SMOTE

# Instantiate the SMOTE instance 
# Set the sampling_strategy parameter equal to auto
smote_sampler = SMOTE(random_state=1, sampling_strategy='auto')
```

```python
X_resampled, y_resampled = smote_sampler.fit_resample(X_train, y_train)
```

```python
y_resampled.value_counts()
```

```python
smote_model = GradientBoostingClassifier(random_state=1)
smote_model.fit(X_resampled, y_resampled)
```

```python
# Predict labels for resampled testing features
smote_y_pred = smote_model.predict(X_test)
```

```python
# Print classification reports
print(f"Classification Report - Original Data")
print(classification_report(y_test, prediction_clf1))
print("---------")
print(f"Classification Report - Resampled Data - SMOTE")
print(classification_report(y_test, smote_y_pred))
```
