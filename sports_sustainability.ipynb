```python
random_tuned_model = GradientBoostingClassifier()
```

```python
X = df.drop('Sustainability Score', axis=1)
y = df['Sustainability Score']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
```

```python
# Create the randomized search estimator
from sklearn.model_selection import RandomizedSearchCV

param_grid =  {'n_estimators': np.arange(100, 1001, 50),
    'learning_rate': np.arange(0.05, 0.31, 0.05),
    'max_depth': np.arange(3, 10),
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'subsample': [0.5, 0.7, 1.0],
    'max_features': ['log2', 'sqrt']}

random_clf = RandomizedSearchCV(random_tuned_model, param_grid, scoring='balanced_accuracy', random_state=0, verbose=3)
```

```python
# Fit the model by using the grid search estimator.
# This will take the KNN model and try each combination of parameters.
random_clf.fit(X_train, y_train)
```

```python
# List the best parameters for this dataset
print(random_clf.best_params_)
```

```python
# Print the classification report for the best model
grid_y_pred = random_clf.predict(X_test)
print(classification_report(y_test, grid_y_pred))
```

```python
# take a look at the balanced accuracy score for the tuned model
balanced_accuracy_score(y_test, grid_y_pred)
```
